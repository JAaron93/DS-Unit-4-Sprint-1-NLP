## Part 1: Tokenize Function
<a id="#p1"></a>

Complete the function `tokenize`. Your function should
- accept one document at a time
- return a list of tokens

You are free to use any method you have learned this week.

````python
# Optional: Consider using spaCy in your function. The spaCy library can be imported by running this cell.
# A pre-trained model (en_core_web_sm) has been made available to you in the CodeGrade container.
# If you DON'T need use the en_core_web_sm model, you can comment it out below.
import spacy
nlp = spacy.load('en_core_web_sm')
````

````python
# Import for regex in a later function
import re
# Increasing the pandas df display for easy viewing
pd.options.display.max_colwidth = 750
df
````

````python
def tokenize(doc):
# YOUR CODE HERE
    
    tokens = []
    
    doc = nlp(doc)
    
    for token in doc:
        # Filtering out punctuation, and stop words. Filtering in lemmas and case normalizion.
        if ((token.is_punct != True) and
            (token.is_lower != True) and
            (token.lemma_ != 'PRON-') and
            (token.is_stop == False)):
                tokens.append(token.lemma_)

    return tokens
````

````python
# Initializing pandarallel to speed up the tokenization process. Applying all 16 vCPUs at my disposal.
from pandarallel import pandarallel
pandarallel.initialize(progress_bar=True, nb_workers=16)
````

````python
%%time
df['tokens'] = df['text'].parallel_apply(tokenize)
````

````python
df
````

````python
# Need to convert this tokens from a list to string for our vectorizer model later on. Will get errors if we don't.
def listintostring(string):
    '''
    This function converts lists into strings
    '''
    str = " "
    return (str.join(string))
````

````python
df['token_list'] = df['tokens'].apply(listintostring)
````

## Part 2: Vector Representation
<a id="#p2"></a>
1. Create a vector representation of the reviews (i.e. create a doc-term matrix).
2. Write a fake review and query for the 10 most similar reviews, print the text of the reviews. Do you notice any patterns?

````python
%%time
# Create a vector representation of the reviews 
# Name that doc-term matrix "dtm"

# YOUR CODE HERE
from sklearn.feature_extraction.text import TfidfVectorizer
# Tuning our vectorizer model
vect = TfidfVectorizer(
                       stop_words='english',
                       ngram_range=(1,2),    # Allowing for both unigrams AND bigrams
                       max_features=10000    # Not allowing more than 10k features/dimensions in our model
                       )
````

````python
# Fitting our model and assigning its output as a DTM to an appropriate name
dtm = vect.fit_transform(df['token_list'])
````

````python
# Converting our DTM to a pandas DataFrame for convenience
dtm = pd.DataFrame(dtm.toarray(), columns=vect.get_feature_names())
dtm
````

````python
# Create and fit a NearestNeighbors model named "nn"
from sklearn.neighbors import NearestNeighbors

# YOUR CODE HERE
nn = NearestNeighbors(n_neighbors=10, algorithm='ball_tree')  # Using ball_tree to measure distance of points
nn.fit(dtm)  # Fitting our DTM to our KNN model
````

````python
# sample a doc from dtm to use as our query point
doc_index = 0
# using integer location(iloc) on the document index in order to sample them
doc = [dtm.iloc[doc_index].values]

# Query Using kneighbors 
neigh_dist, neigh_index = nn.kneighbors(doc)
````

````python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Name the gridsearch instance "gs"

# YOUR CODE HERE
# Using tfidf vect and KNN for our models' pipeline
vector = TfidfVectorizer()
knnc = KNeighborsClassifier(n_jobs=-1)

pipe = Pipeline([('vect', vector),
                 ('clf', knnc)])
# Doing some statistical trimming with max_df to our vector. Applying distance and uniformity based weights to our knn
params = {
          "vect__max_df": (0.5, 0.95),
          "clf__weights": ['distance', 'uniform']
          }

gs = GridSearchCV(
                  pipe,
                  params,
                  n_jobs=-1,       # Using all our CPU resources on this
                  cv=5,            # 5 cross validation folds where fifths of our data are trained and validated on each other
                  verbose=2
                  )

gs.fit(X, y)
````

````python
# Predicting with our trained, gridsearch tuned model
predict = gs.predict(fake_review)
list(predict)
````
